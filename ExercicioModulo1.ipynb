***Calculadora de IMC - Índice de Massa***

```python
peso = input("Digite seu peso") #Aguardar o usuário digitar seu peso
print(peso)
```

```python
peso_convert = float (peso) #converti o valor do peso em valor flutuante
print(peso_convert)
```

```python
type(peso_convert) #verificando o tipo da variável
```

```python
altura = input("Digite sua altura") #Aguardar o usuário digitar sua altura
print(altura)
```

```python
altura_convert = float (altura) #converti o valor da altura em valor flutuante
print(altura_convert)
```

```python
type(altura_convert) #verificando o tipo da variável
```

```python
IMC = peso_convert / (altura_convert**2) #Calculo IMC
print('Seu IMC é de', IMC)
```

```python
type(IMC)
```

```python
# Calculo tabela IMC
if IMC >= 18.5 and IMC <= 24.9:
  print("Você está com o Peso normal")
elif IMC >=25 and IMC <= 29.9:
  print("Você está com Sobrepeso")
elif IMC > 30:
  print("Você está com Obesidade")
else:
  print("Erro")
```

```python
#Definir o peso ideal
Peso_ideal = 24.9 * (altura_convert**2)
print('Seu peso ideal é', Peso_ideal)
```

```python
#Quantos Kg precisa perder
Kg_perder = peso_convert - Peso_ideal
print('Você precisa perder', Kg_perder, 'para atingir seu peso normal')
```

```python
#Média de tempo necessários para perder o excesso de peso - 0.500 g por semana
Tempo_perda_peso = 0.5 * Kg_perder
print('Vai eliminar o excesso de peso em', Tempo_perda_peso, "meses")
```

```python
#Lista de Atividades que ajudam na perda de peso - Considerando 1 hora
lista_atividades = [("Corrida", 700), ("Natação", 600), ("Caminhada", 300)]

#Perguntar ao usuário o tempo de pratica por dia
for atividade, calorias_horas in lista_atividades:
  tempo = float (input (f"Quanto tempo você praticará {atividade} (em horas)?"))
  calorias_totais = tempo * calorias_horas
  print(f"Em {atividade}, você queimará aproximadamente {calorias_totais} calorias.")
```
